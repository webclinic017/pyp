3. Numpy的数组与Python传统列表有什么区别

Numpy数组的所有元素必须是同一类型，例如，同时存在整型、浮点型、字符串的数组，Numpy会自动把所有元素转换为字符串，三者的转换优先顺序为：
字符串 > 浮点型 > 整型

5. 分别使用 `linspace` & `arange` 创建一个等差数列

```python
arr1 = np.linspace(0, 100, num=20)
arr1
```

```python
arr2 = np.arange(100, step=2)
arr2
```
