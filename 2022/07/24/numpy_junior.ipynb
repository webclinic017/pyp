3. Numpy的数组与Python传统列表有什么区别

1. np的数组所有元素必须是同一类型，若存在不同类型，会自动转换为同一类型。例如元素类型有整型、字符串、浮点型，会统一转换为字符串，三者优先级为 字符串 > 浮点型 > 整型，可看出规律：空间占用小的类型向空间占用大的类型转换。
2. np数组创建的时候具有固定大小，python原生数组可动态增长。更改ndarray的大小将创建一个新的数组并删除原来的数组。
3. np数组对大量数据进行高级数学运行非常友好，对比python原生数组执行效率更高，且编写的代码量更少。
