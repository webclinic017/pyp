18. 级联操作：将多个numpy数组进行横向或纵向拼接。(特殊场景：行列个数不一样)

20. 创建一个 `size=(5, 6)` 的随机数组，分别求出每一行、每一列的sum、max

```python
n20 = np.random.randint(0, 100, size=(5, 6))
print(n20)
print("----------------------")
print("sum:", n20.sum(axis=1))
print("max: ", n20.max(axis=0))
```

21. 计算标准差、方差

```python
# n20 = np.ones(shape=(5, 6))
print("mean:", n20.mean(axis=1))
print("var:", n20.var(axis=1))
print("std:", n20.std(axis=1))
```

22. 计算数组中元素最大值与最小值的差

```python
print(n20)
print("--------------")
print("ptp:", n20.ptp(axis=1))
```
